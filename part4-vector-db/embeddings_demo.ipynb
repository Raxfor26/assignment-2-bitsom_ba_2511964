# Part 4 - Vector Databases: Embeddings & Similarity

This notebook demonstrates:
- Generating sentence embeddings using `sentence-transformers` (`all-MiniLM-L6-v2`)
- Computing a 30x30 cosine similarity matrix visualized as a heatmap
- Finding the top-2 most similar sentences for a new query

## Step 1: Install Dependencies

```python
import sys
import subprocess
subprocess.check_call([sys.executable, '-m', 'pip', 'install',
    'sentence-transformers', 'scikit-learn', 'matplotlib', 'seaborn', '--quiet'])
print('Dependencies installed.')
```

## Step 2: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

print('All libraries imported successfully.')
```

## Step 3: Define 30 Sentences Across 3 Topics (10 each)

```python
sentences = [
    # Cricket (10 sentences)
    'The batsman hit a magnificent six over the boundary.',
    'India won the test match by an innings and fifty runs.',
    'The spinner deceived the batsman with a googly on a dusty pitch.',
    'The fielder took a stunning catch at deep square leg.',
    'The bowler delivered a perfect yorker to clean bowl the opener.',
    'The captain declared the innings after reaching a commanding total.',
    'A dropped catch in the slips cost the team a crucial wicket.',
    'The umpire raised his finger to signal the batsman was leg before wicket.',
    'The opening partnership put on over a hundred runs before the first wicket fell.',
    'Rain interrupted play on the third day of the test match.',

    # Cooking (10 sentences)
    'Saute the onions in olive oil until they turn golden brown.',
    'Always preheat the oven before baking bread or pastries.',
    'Marinating the chicken overnight enhances the flavor significantly.',
    'Fold the egg whites gently into the batter to keep it light and airy.',
    'Simmer the sauce on low heat for thirty minutes to develop deep flavor.',
    'Use a sharp knife to julienne the vegetables into thin uniform strips.',
    'Blanching the vegetables preserves their color and reduces bitterness.',
    'Deglaze the pan with white wine to lift the caramelized bits from the bottom.',
    'Let the dough rest for an hour before rolling it out on a floured surface.',
    'Season the pasta water generously with salt before adding the noodles.',

    # Cybersecurity (10 sentences)
    'A firewall monitors and controls incoming and outgoing network traffic.',
    'Phishing attacks trick users into revealing sensitive credentials.',
    'Encrypting data ensures it remains unreadable to unauthorized parties.',
    'Two-factor authentication adds an extra layer of security beyond passwords.',
    'A zero-day vulnerability is exploited before the vendor releases a patch.',
    'Penetration testers simulate real attacks to identify system weaknesses.',
    'Ransomware encrypts victim files and demands payment for the decryption key.',
    'Social engineering manipulates individuals into divulging confidential information.',
    'A VPN tunnels internet traffic through an encrypted connection to protect privacy.',
    'Regular software updates are essential to patch known security vulnerabilities.'
]

labels = [
    'Cricket-1',  'Cricket-2',  'Cricket-3',  'Cricket-4',  'Cricket-5',
    'Cricket-6',  'Cricket-7',  'Cricket-8',  'Cricket-9',  'Cricket-10',
    'Cooking-1',  'Cooking-2',  'Cooking-3',  'Cooking-4',  'Cooking-5',
    'Cooking-6',  'Cooking-7',  'Cooking-8',  'Cooking-9',  'Cooking-10',
    'Cyber-1',    'Cyber-2',    'Cyber-3',    'Cyber-4',    'Cyber-5',
    'Cyber-6',    'Cyber-7',    'Cyber-8',    'Cyber-9',    'Cyber-10'
]

print(f'Total sentences: {len(sentences)}')
for label, sent in zip(labels, sentences):
    print(f'  [{label}] {sent}')
```

## Step 4: Load Model and Generate Embeddings

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)

print(f'Embedding shape: {embeddings.shape}')
print('Each sentence is represented as a 384-dimensional vector.')
```

## Step 5: Compute 30x30 Cosine Similarity Matrix

```python
sim_matrix = cosine_similarity(embeddings)

print('Cosine Similarity Matrix (30x30):')
print(np.round(sim_matrix, 3))
```

## Step 6: Visualize as Heatmap

```python
plt.figure(figsize=(18, 15))
sns.heatmap(
    sim_matrix,
    xticklabels=labels,
    yticklabels=labels,
    annot=True,
    fmt='.2f',
    cmap='YlOrRd',
    vmin=0,
    vmax=1,
    linewidths=0.3,
    linecolor='gray',
    annot_kws={'size': 6}
)
plt.title('30x30 Cosine Similarity Matrix\n(Cricket | Cooking | Cybersecurity)', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=7)
plt.yticks(rotation=0, fontsize=7)
plt.tight_layout()
plt.savefig('similarity_heatmap.png', dpi=150)
plt.show()
print('Heatmap saved as similarity_heatmap.png')
```

## Step 7: Query - Find Top-2 Most Similar Sentences

**Query:** *"The bowler took three wickets in one over"*

```python
query = 'The bowler took three wickets in one over'
query_embedding = model.encode([query])

query_similarities = cosine_similarity(query_embedding, embeddings)[0]

top2_indices = np.argsort(query_similarities)[::-1][:2]

print(f'Query: "{query}"')
print('-' * 60)
print('Top 2 Most Similar Sentences:')
for rank, idx in enumerate(top2_indices, start=1):
    print(f'  Rank {rank}: [{labels[idx]}] {sentences[idx]}')
    print(f'           Similarity Score: {query_similarities[idx]:.4f}')
```
